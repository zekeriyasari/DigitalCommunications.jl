# Monte Carlo Simulations of Digital Communication Systems

##  Introduction 

Consider the baseband equivalent model of a communication system shown below. 

![waveformblockdiagram](../assets/tikz/bandpasswaveformdiagram.svg)

`Modulator` maps each symbol $m_i$ into a waveform $s_i(t), \; i = 1, \ldots, M = 2^k$ corresponding to its scheme (eg. PAM, ASK, PSK, QAM, FSK). The waveform $s_i(t)$ is corrupted by an additive white Gaussian noise $n_i(t)$ which is asssumed to be zero mean and to have a constant power spectral density of $N_0/2$. `Detector` recovers the trasmitted symbol $m_i$ as $\hat{m}_i$. Due to the corruption in the channel, errornous detection occurs in the detector. This errornous detection is quantified as the probability of symbol error $P_e$ in terms of the channel noise level given as the energy per symbol to noise power spectral density ratio $\dfrac{E_s}{N_0}$.


## Baseband and Bandpass Signals 

A signal $s(t)$ is a bandpass signal if its double side band magnitude spectrum $S(f)$ is centered ($S(f) \neq 0$) around the frequencies $-f_c$ and $f_c$ within a band of frequencies $[-f_c - B, -f_c + B]$ and $[f_c - B, f_c + B]$. $f_c$ is called as the center frequency and $2B$ is the bandwidth of $s(t)$. A signal $s_l(t)$ is called as a baseband (lowpass) signal its double side band magnitude spectrum is centered around the origin, i.e., $S_l(f) \neq 0$ for $f \in [-B, B]$ where $B$ is the bandwidth. A bandpass $s(t)$  can be represented by a baseband signal $s_l(t)$ and vice versa. The equation below show the relation between a bandpass signal and its baseband equivalent. 
$$
\begin{aligned}
    s(t) &= \Re[s_l(t)e^{j 2\pi f_c t}] \\
    s_l(t) &= (s(t) + j \hat{s}(t)) e^{-j 2\pi f_c t} \\ 
    S(f) &= \dfrac{1}{2}(S_l(f + f_c) + S_l^*(-f - fc)) \\ 
    S_l(f) &= 2 S(f + f_c) u(f + f_c) 
\end{aligned}
$$  
where $f_c$ is the center frequency of $s(t)$ and $\hat{s}(t)$ is the Hilbert transform of $s(t)$ defined as $\hat{s}(t) = s(t) * \dfrac{1}{\pi t}$, where $x(t)*y(t)$ is the convolution of $x(t)$ and $y(t)$ defined as 
$$
x(t) * y(t) = \int_{-\infty}^{\infty} x(\tau) y(t - \tau) d\tau
$$

## Energy, Power and Expansion of Signals 

The signal eneryg $\mathcal{E}_x$ of the signal $x(t)$ is defined as 
$$
\mathcal{E}_x = \int_{-\infty}^{\infty} |x(t)|^2 dt
$$
where $|x(t)|$ denotes the absolute of the signal $x(t)$. The power $\mathcal{P}_x$ averaged for a time duration $T$ of $x(t)$ is defined as 
$$
\mathcal{P}_x = \lim_{t \mapsto \infty} \dfrac{1}{T} \int_{-T/2}^{T/2} |x(t)|^2 dt 
$$
$x(t)$ is called as energy signal if $0 < \mathcal{E}_x < \infty $ and is called as power signal if $0 < \mathcal{P}_x < \infty$ 

The inner product of the signals $x(t)$ and $(y(t)$ is defined as 
$$
s_j = \langle s(t), \phi(t) \rangle = \int_{-\infty}^{\infty} s(t) \phi_j(t)dt 
$$
Note that $E_x = \langle x(t), x(t) \rangle$.  It can be shown that 
$$
\begin{aligned} 
\mathcal{E}_{s} = \dfrac{1}{2}\mathcal{E}_{sl} 
\end{aligned}
$$
where $s$ is the baseband signal with lowpass equivalent signal $s_l$.

$x(t)$ and $y(t)$ is orthogonal if their inner products is zero. The set of signals $\{\phi_j(t) \}_{j = 1}^{N}$ is orthogonal if  
$$
\langle \phi_i(t), \phi_j(t) \rangle = 
\begin{cases} 
\rho \neq 0 & i = j \\
0 & i \neq j
\end{cases}
$$
The set of signals $\{\phi_j(t) \}_{j = 1}^N$ is orthonormal if $E_{\phi_i} = 1$ for $i = 1, \ldots, N$

If there exists an orthonormal set of signals $\{\phi_j(t) \}$, then a signal $s(t)$ can be expanded in terms of the signals in the orthonormal set 
$$
s(t) = \sum_{j = 1}^N s_j \phi_j(t)
$$
where 
$$
s_i = \langle s(t), \phi_j(t)$.
$$

## Digital Modulation Schemes 

Modulation schemes determines how the symbols are mapped into waveforms to be transmitted through the channel. There different types of digital modulation schemes. Some of them are summarized below. 

### PAM 
In the pulse amplitude modulation, we signal transmitted is defined as 
$$
s_m(t) = \Re[A_m g(t)] = A_m g(t)  \quad A_m = 2m - 1 - M \quad m = 1, \ldots, M
$$
where $g(t)$ is the modulation pulse defined over a time interval of $[0, T]$, where $T$ is the symbol duration. Note that PAM is a baseband modulation type. If we choose $\{ \phi(t)\}$ where 
$$
\phi(t) = \dfrac{g(t)}{\sqrt{\mathcal{E}_g}}
$$
we can express $s_m(t)$ as a vector with a single element as follows.
$$
s_m = [A_m \sqrt{\mathcal{E}_g}]
$$

### ASK 
Amplitude shift modulation is very similar to the PAM. $s_m(t)$ is defined as 
$$
s_m(t) = \Re[A_m g(t) e^{j 2\pi f_c t}] = A_m g(t) cos(2 \pi f_c t)  \quad A_m = 2m - 1 - M \quad m = 1, \ldots, M
$$
where $g(t)$ is the modulation pulse defined over a time interval of $[0, T]$, where $T$ is the symbol duration. If $\{\phi(t)\}$ is chosen as a basis where 
$$
\phi(t) = \sqrt{\dfrac{2}{\mathcal{E}_g}} g(t) \cos(2 \pi f_c t)

$$
we can express $s_m(t)$ as a vector with a single element as follows.
$$
s_m = [A_m \sqrt{\dfrac{\mathcal{E}_g}{2}}]
$$

Note that the baseband equivalent for $s_m(t)$ is 
$$
s_{ml} = A_m g(t)
$$
If we choose the set of signal ${\phi(t)}$ such that 
$$
\phi(t) = \dfrac{g(t)}{\sqrt{\mathcal{E}_g}}
$$,
the single element vector 
$$
s_{ml} = [A_m \sqrt{\mathcal{E}_g}]
$$
can represent $s_{ml}(t)$.

### PSK
In phase shift modulation, the information is burried into the phase of the carrier. $s_m(t)$ is defined as 
$$
s_m(t) = \Re[e^{j \theta_m} g(t) e^{j 2\pi f_c t}] = g(t) cos(2 \pi f_c t + \theta_m)  \quad \theta_m = \dfrac{2\pi (m - 1)}{M} \quad m = 1, \ldots, M
$$
where $g(t)$ is the modulation pulse defined over a time interval of $[0, T]$, where $T$ is the symbol duration. If $\{\phi_1(t), \phi_2(t)\}$ is chosen as a basis where 
$$
\begin{aligned}
\phi_1(t) &= \sqrt{\dfrac{2}{\mathcal{E}_g}} g(t) \cos(2 \pi f_c t) \\
\phi_2(t) &= \sqrt{\dfrac{2}{\mathcal{E}_g}} g(t) \cos(2 \pi f_c t)
\end{aligned}
$$
we can express $s_m(t)$ as a vector with two elements as follows.
$$
s_m = [cos(\theta_m) \sqrt{\dfrac{\mathcal{E}_g}{2}}, sin(\theta_m) \sqrt{\dfrac{\mathcal{E}_g}{2}}]
$$

The baseband equivalent for $s_m(t)$ is 
$$
s_{ml} = A_m g(t)
$$
where $A_m = cos(\theta_m) + j sin(\theta_m)$ is  a complex number If we choose the set of signal ${\phi(t)}$ such that 
$$
\phi(t) = \dfrac{g(t)}{\sqrt{\mathcal{E}_g}}
$$,
the complex vector with a single element 
$$
s_{ml} = [A_m \sqrt{\mathcal{E}_g}]
$$
can represent $s_{ml}(t)$.

### QAM
In quadrature amplitude modulation, the information is burried both into the phase and amplituded of the carrier. $s_m(t)$ is defined as 
$$
s_m(t) = \Re[A_m g(t) e^{j 2\pi f_c t}] = A_{mi} g(t) cos(2 \pi f_c t) - A_{mq} g(t) sin(2 \pi f_c t)  \quad A_m = A_{mi} + j A_{mq} \quad m = 1, \ldots, M
$$
where $g(t)$ is the modulation pulse defined over a time interval of $[0, T]$, where $T$ is the symbol duration. If $\{\phi_1(t), \phi_2(t)\}$ is chosen as a basis where 
$$
\begin{aligned}
\phi_1(t) &= \sqrt{\dfrac{2}{\mathcal{E}_g}} g(t) \cos(2 \pi f_c t) \\
\phi_2(t) &= -\sqrt{\dfrac{2}{\mathcal{E}_g}} g(t) \sin(2 \pi f_c t)
\end{aligned}
$$
we can express $s_m(t)$ as a vector with two elements as follows.
$$
s_m = [A_{mi} \sqrt{\dfrac{\mathcal{E}_g}{2}}, A_{mq} \sqrt{\dfrac{\mathcal{E}_g}{2}}]
$$

The baseband equivalent for $s_m(t)$ is 
$$
s_{ml} = A_m g(t)
$$
where $A_m = SA_{mi} + A_{mq}$ is a complex number. If we choose the set of signal ${\phi(t)}$ such that 
$$
\phi(t) = \dfrac{g(t)}{\sqrt{\mathcal{E}_g}}
$$,
the complex vector with a single element 
$$
s_{ml} = [A_m \sqrt{\mathcal{E}_g}]
$$
can represent $s_{ml}(t)$.

### FSK
In frequency shift keying modulation, the information is burried both into the frequency of the carrier. $s_m(t)$ is defined as 
$$
s_m(t) = \Re[e{j 2 \pi m \Delta f t} g(t) e^{j 2\pi f_c t}] = g(t) cos(2 \pi (f_c + m \Delta f) t)  \quad m = 1, \ldots, M
$$
where
$$
g(t) = 
\begin{cases}
\sqrt{2 \mathcal{E}_s}{T} & 0 \leq t \leq T \\
0 & 0 otherwise \\
\end{cases} 
$$
where $T$ is the symbol duration.

If $\{\phi_j(t)\}_{j = 1}^{M}$ is chosen as a basis where 
$$
\phi_j(t) = \dfrac{s_j(t)}{\sqrt{\mathcal{E}_s}}
$$
we can express $s_m(t)$ as a vector with $M$ elements as follows.
$$
\begin{aligned}
s_1 &= [\sqrt{\mathcal{E}_s}, 0, 0, \ldots, 0, 0]  \\  
s_2 &= [0, \sqrt{\mathcal{E}_s}, 0, \ldots, 0, 0] \\  
\vdots &= \vdots \\
s_M &= [0, 0, 0, \ldots, 0,  \sqrt{\mathcal{E}_s}] \\  
\end{aligned}
$$

The baseband equivalent for $s_m(t)$ is 
$$
s_{ml} = e^{j 2 \pi \Delta f t} g(t)
$$
If we choose the set of signal $\{\phi_{jl}(t)\}_{j = 1}^{M}$ such that 
$$
\phi_{jl}(t) = \dfrac{e^{j 2 \pi \Delta f t} g(t)}{\sqrt{2}}
$$,
the vectors $s_{jl}$ can represent $s_{jl}(t)$ where
$$
\begin{aligned}
s_{1l} &= [\sqrt{2\mathcal{E}_s}, 0, 0, \ldots, 0, 0]  \\  
s_{2l} &= [0, \sqrt{2\mathcal{E}_s}, 0, \ldots, 0, 0] \\  
\vdots &= \vdots \\
s_{Ml} &= [0, 0, 0, \ldots, 0,  \sqrt{2\mathcal{E}_s}] \\  
\end{aligned}
$$

## Baseband and Bandpass Random Processs

The autocorrelation function of a random process $X(t)$ is define as 
$$
R_X(\tau) = E[X(t) X(t + \tau)]
$$
where $E[X]$ denotes the expectation of a random variable $X$. For continous wide sense stationary random process $X(t)$, we have, 
$$
    R_X(\tau) \underset{\mathcal{F}^{-1}}{\stackrel{\mathcal{F}}{\rightleftharpoons}} S_X(f) 
$$
where $\mathcal{F}[x(t)]$ denotes the Fourier transform of $x(t)$ defined as 
$$
\mathcal{F}[x(t)] = \int_{-\infty}^{\infty} x(t) e^{-j 2 \pi f t}dt 
$$
and $\mathcal{F}^{-1}[X(f)]$ is the inverse Fourier transform of $X(f)$ defined as 
$$
\mathcal{F}^{-1}[X(f)] = \int_{-\infty}^{\infty} X(f) e^{-j 2 \pi f t}df 
$$

$S_X(f)$ is called as the power spectral density of $X(t)$. For we have, 
$$
\sigma_X^2 = R_X(0) = \int_{-\infty}^{\infty} S_n(f) df = \mathcal{P}_X
$$

A random process $X(t)$ is a called bandpass process if its autocorrelation function $R_X(\tau)$ is a bandpass signal. Similarly, $X(t)$ is called baseband process if its autocorrelation function $R_X(\tau)$ is a baseband signal. 

Similar to the case of deterministic signals, a bandpass random process $X(t)$ can be represented by a baseband random process $X_l(t)$ and vice versa. The relations between $S_{Xl}(f)$ and $S_X(f)$ given below summarizes the relation between $X(t)$ and $X_l(t)$. 
$$
\begin{aligned}
S_{Xl}(f) &= 
\begin{cases} 
4 S_X(f) & f \in [-B, B] \\ 
0 & otherwise 
\end{cases} \\ 
S_{X} &= \dfrac{1}{4}(S_{Xl}(f + f_c) + S_{Xl}(-f - f_c))
\end{aligned}
$$
where $B$ is the bandwidth of the random process $X_l(t)$ and $f_c$ is the central frequency of $X(t)$. From this relation, if a zero mean white random process $n(t)$ with constant power spectral density $N_0/2$ is filtered by a bandpass filter with center frequency $f_c$ and bandwidth $2B$, then its baseband equivalent $n_l(t)$ has a power spectral density of $2N_0$ for the frequency range of $[-B, B]$, i.e., 
$$
S_{nl}(f) = 
\begin{cases}
2N_0 & f \in [-B, B] \\
0 & otherwise \\
\end{cases} 
$$
The baseband noise $n_l(t)$ is complex valued and in the form $n_l(t) = n_{li}(t) + j n_{lq}(t)$. Since $R_{nl}(t) = \mathcal{F}^{-1}(S_{nl}(f))$, we have, 
$$
R_{nl}(\tau)  = 2 N_0 B sinc(2 B \tau)
$$
The variance of the lowpass signal can be found, 
$$
\sigma_{nl}^2 = 2 N_0 B
$$
Since $n_l(t)$ is a complex random process its variance is distributed equally to its real and imaginary parts, that is, $\sigma_{ni}^2 = N_0 = \sigma_{lq}^2$.

Let us define a discrete time noise $n_l[k] = n_l(k T_s), \; k = 1, 2, \ldots$ where $T_s$ is the sampling frequency. We want the samples of $n_l[k]$ to be uncorrelated, that is, $R_{nl}[k] = \delta[k]$ where $\delta[k]$ is the impulse function defined as 
$$
\delta[k] = 
\begin{cases} 
1 & k = 0 \\ 
0 & otherwise 
\end{cases}
$$
Let $t = k T_s$ and $\tau = (k + m) T_s$. Then,
$$
R_{nl}[k] = E[n_l[k] n_l[k + m]] = E[n_l(k T_s) n_l((k + m)T_s)] = 2 N_0 B sinc(2 B m T_s)
$$
If the sampling frequency $f_s = 1 / T_s$ is chosen such that 
$$
f_s = 2B / m \quad m = 1, 2, \ldots 
$$
the noise sample $n_l[k]$ becomes uncorrelated, and hence white. Finally the variance of the sampled noise $n_l[k]$ is $\text{var}[n_l[k]] = 2 N_0$

## Calculation of Variance 
Generally, noise level of a communication system is given in terms of $\dfrac{\mathcal{E}_s}{N_0}$ where $\mathcal{E}_s$ is the average energy per symbol and $N_0$ is the noise power spectral density of the additive white Gaussian noise. This quantity is defined for the signals in the passband. However the communiation system to be simulated here operates in the baseband and all the signals flowing through the communication system are their sampled version, hence are in discrete time. We use random number generator tools such as `randn` and so we can just specify the variance of the sampled noise process. Therefore, we must calculate correctly the variance of the sampled bandpass noise process. 

One $\dfrac{\mathcal{E}_s}{N_0}$ is defined, (possible in units of dB, if so we must convert dB scale to normal scale), we can write, 
$$
\left(\dfrac{\mathcal{E}_s}{N_0} \right) = \dfrac{\mathcal{E}_s}{N_0} 
$$
where the quantity $\left(\dfrac{\mathcal{E}_s}{N_0} \right)$ denotes the value calcuate from the $\dfrac{\mathcal{E}_s}{N_0} $ given in dB scale. The conversion is performed as 
$$
\left(\dfrac{\mathcal{E}_s}{N_0} \right) = 10^{\dfrac{\mathcal{E}_s}{N_0} / 10}
$$
Then, the average signal energy $\mathcal{E}_s$ must be calculated from the input sequence of the channel
$$
\mathcal{E}_s = \dfrac{1}{K} \sum_{k = 1}^K \mathcal{E}_{sk} =  \dfrac{1}{K} \sum_{k = 1}^K |s_k|^2
$$
where $s_k$ is the $k^{th}$ symbol (vector of expansion coefficients or waveform). $\mathcal{E}_{sk}$ is the energy of the bandpass symbol $k^{th}$ transmitted and this signal is a bandpass signal. However, the communication being simulated operates in the passband and we will compute the average symbol in energy in the baseband. We know that 
$$
\mathcal{E}_{sk} = \dfrac{1}{2}\mathcal{E}_{skl}
$$
where $\mathcal{E}_{skl}$ is the energy of the baseband equivalent signal $s_{kl}$ of the bandpass signal $s_k$. This implies 
$$
\mathcal{E}_{sl} = \dfrac{1}{K} \sum_{k = 1}^K \mathcal{E}_{skl} =  \dfrac{1}{K} \sum_{k = 1}^K |s_{kl}|^2 = 2 \mathcal{E}_{s}
$$
Thus, we have, 
$$
\left( \dfrac{\mathcal{E}_s}{N_0} \right) = \dfrac{\mathcal{E}_{s}}{N_0} = \dfrac{\mathcal{E}_{sl}}{2N_0} \quad \Rightarrow \quad  2N_0 = \dfrac{\mathcal{E}_{sl}}{\left( \dfrac{\mathcal{E}_s}{N_0} \right)}
$$

In the previous section, it was show that the variance of the sampled noise $n_l[k]$ is $2N_0 f_s$ where $f_s$ is the sampling frequency. Hence 
$$
\text{var}[n_l[k]] = \dfrac{\mathcal{E}_{sl}}{\left( \dfrac{\mathcal{E}_s}{N_0} \right)} f_s
$$

## Detection

The optimum decision rule is given as 
$$
\hat{m} = \arg\max_{1 \leq m \leq M}[N_0 ln P_m - \dfrac{1}{2}\mathcal{E}_{ml} + r_l \cdot s_{ml}]
$$
where $P_m$ is the probility that the symbol $m$ is transmitted, $\mathcal{E}_{ml}$ is the energy of the lowpass signal $s_{ml}(t)$, $r_l$ and $s_{ml}$ are the vector representations of the received lowpass signal $r_l(t)$ and $s_{ml}(t)$, respectively.   

## Simulation 

In the baseband equivalent models, the baseband symbol waveforms are in the form 
$$
s_{ij}(t) = \sum_{j = 1}^{N} s_{ij} \phi_{jl}(t)
$$
where $\{\phi_{jl}(t)\}_{j = 1}^{N}$ is the basis of the modulation. 

In the following subsections we will perform Monte Carlo simulations of 8-PSK scheme using two different approach. In the first approach vector channel model in which the modulator transmits just vectors $s_i = [s_{i1}, s_{i2}, \ldots, s_{iN}]$ where $s_{ij}$'s are the coefficients of the expansion of the baseband signal $s_{il}(t) \in {s_{ml}(t)}_{m = 1}^{M}$ over the basis $\{\phi_{jl}(t)\}_{j = 1}^{N}$. On the other hand, in the waveform channel approach, sampled baseband waveform $s_{il}[k] = s_{il}(k T_s)$, where $T_s$ is the sampling period, is transmitted.


### Waveform Channel Model 

Let us first load the required packages and determines the settings 

```python
# MonteCarlo simulation of PSK modulation scheme using the waveform channel model 

using DigitalCommunications 
using Plots 

# Settings
pulse = RectangularPulse()          # Modulation pulse 
fs = 10                             # Samling frequency 
ts = 1 / fs                         # Sampling period 
tb = pulse.duration                 # Pulse duration 
k = 3                               # Bits per symbol 
M = 2^k                             # Costellation size 
nsymbols = Int(1e6)                 # Number of symbols 
ebno = collect(0 : 10)              # Energy per bit to noise power spectral density ratio in dB 
esno = ebno .+ 10 * log10(k);       # Energy per symbol to noise power spectral density ratio in dB
```

Note that we are simulating 8-PSK system for an $\dfrac{\mathcal{E}_s}{N_0}$ range of $[0, 10]$ for $1$ dB resolution.  We will use a rectangular pulse. Let us investigate the pulse a little bit more. 

```python
pulse.duration, pulse.amplitude
```

The amplitude of the pulse is 1 and the pulse duration, i.e., the symbol duration is 1 seconds. In other words, symbol transmission rate R = 1 symbols/second. The waveform of the pulse is plotted next. 

```python
ti = -pulse.duration            # Initial time 
dt = pulse.duration / 100       # Sampling period 
tf = 2 * pulse.duration         # Final time 
tv = ti : dt : tf 
plot(tv, pulse.(tv))
xlabel!("time [seconds]") 
ylabel!("g(t)")
title!("Modulation pulse")
```

As expected the pulse energy is 1.

```python
energy(pulse)
```

We next construct the blocks of the communcation system. For this case, we will simulate the communication system usign the waveform channel model. To this end, let us construct a symbol generator, a modulator whose scheme is 8-PSK, pulse is rectangular `pulse` and sampling period is 'ts` defined above.  The channel is an additive white Gaussian noise channel with $esno = 1$, sampling period $ts$ and symbol duration of $tb$ seconds. Finally, we need a detector whose reference are calculated according the modulator. 

```python
# Communication system components  
gen = SymbolGenerator(nsymbols, M) 
modulator = Modulator(PSK(M), RectangularPulse(), ts)
channel = AWGNChannel(1., ts, tb)
detector = Detector(modulator(1:M));
```

Before runnning the whole system let us investigate the components a little bit. For example, see the contents of `gen` 

```python
gen.symbols
```

The content of `gen` is the symbols $m_i \in \{ 1, 2, \ldots, 8\}$. The `modulator` can modulate, i.e, maps these symbols into waveforms. Let us see the waveforms for $m = 1, 2, \ldots, 8$ 

```python
refsymbols = 1 : M 
waveforms =  modulator(refsymbols)
```

Let us take a look at the waveform corresponding to the first symbol 

```python
@show refsymbols[1] 
@show waveforms[1]
@show length(waveforms[1]);
```

Note that 10 samples has been taken from the waveform. All the remaining waveforms are sampled with the same number of samples. 

An the type of the `detector` is maximum likelihood (ML), which can be justified as

```python
isml(detector)
```

Let us plot the reference signals of the modulator 

```python
# Plot of basis signals 
ti = 0. 
dt = modulator.pulse.duration / 100 
tf = modulator.pulse.duration 
tv = ti : ts : tf - ts 
plt = plot()
for (k, base) in enumerate(basis(modulator)) 
    plot!(tv, base.(tv), label=k)
end 
xlabel!("time [seconds]") 
ylabel!("g(t)") 
ylims!(-0.2, 1.2)
title!("Basis signals in the baseband")
plt
```

Note that since in we have just a single basis function $\phi(t)$ for PSK scheme in the baseband and this is given as 
$$
\phi(t) = g(t) / \sqrt{\mathcal{E}_g}
$$
where $g(t)$ is the modulation pulse. 

Having investigated the blocks to some extent, we can now run the complete simulation. In the code block given below, we adjust the noise level, i.e `esno`, of the channel for the noise level range set above, corrupt the transmitted symbols in the channel and detect the transmitted `message` in the detector as `mbar`. A symbol error is calculated for each noise level . At last, the numerical error results are compared to theoretically expected values. For the sake of comparison, the theoretical expectations and numerical results are ploted. The plots show that the results are quitte close to each other. 

```python
# Monte Carlo simulation 
message = gen.symbols  
symerr = zeros(length(esno))
for i in 1 : length(symerr)
    channel.esno = esno[i]  # Update channel noise level 
    mbar = message |> modulator |> channel |> detector  # Run communication system 
    symerr[i] = sum(mbar .!= message) / length(message)  # Symbol error rate 
end

# Plots
plt = plot(title="$M-PSK", xlabel="ebno [dB]", ylabel="Pe") 
plot!(ebno, berpsk.(esno, M), marker=:circle, yscale=:log10, label="theoretical")
plot!(ebno, symerr, marker=:circle, yscale=:log10, label="montecarlo")
```

```python
# Monte Carlo simulation 
message = gen.symbols  
symerr = zeros(length(esno))
for i in 1 : length(symerr)
    channel.esno = esno[i]  # Update channel snr
    mbar = message |> modulator |> channel |> detector  # Extracted message signal 
    symerr[i] = sum(mbar .!= message) / length(message)  # Symbol error rate 
end

# Plots
plt = plot(title="$M-PSK", xlabel="ebno [dB]", ylabel="Pe") 
plot!(ebno, berpsk.(esno, M), marker=:circle, yscale=:log10, label="theoretical")
plot!(ebno, symerr, marker=:circle, yscale=:log10, label="montecarlo")
```
